```python
# from https://stackabuse.com/using-plotly-library-for-interactive-data-visualization-in-python/
import pandas as pd
import numpy as np
%matplotlib inline

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)

import cufflinks
cufflinks.go_offline()
```

```python
import wc_sim
```

```python
sim_run = '/root/wc_repos/wc_sim/runs/2020-12-25/16:48:26.878257'
sim_run = '/root/wc_repos/runs/2020-12-27/14:41:52.567833'
run_results = wc_sim.RunResults(sim_run)
```

```python
pops = run_results.get('populations')
rate_laws = run_results.get('rate_laws')
dfba_reaction_fluxes = run_results.get('dfba_reaction_fluxes')
masses = run_results.get_masses()
volumes = run_results.get_volumes()
```

```python
non_0_trajectories = np.any(pops, axis=0)
species_touch_0 = non_0_trajectories[non_0_trajectories == False].index
print('Species always 0:', ", ".join(sorted(species_touch_0)))
```

```python
last_pops = pops.values[-1, :]    # row, column
last_pops_series = pd.Series(last_pops, pops.columns)
# last_pops_series.head()
species_finish_at_0 = last_pops_series[last_pops_series == 0.].index
print('Species that finish at 0:', ", ".join(sorted(species_finish_at_0)))
```

```python
pops.iplot(layout=dict(xaxis=dict(title='Time'),
                       yaxis=dict(type='log',
                                  title='Population',
                                  range=[1, 5]),
                       title='Populations'))
```

```python
rate_laws.iplot(layout=dict(xaxis=dict(title='Time'),
                       yaxis=dict(type='log',
                                  title='Rate (1/sec)'),
                       title='Rate Laws (ignoring whether reaction is enabled)'))
```

```python
dfba_reaction_fluxes.iplot(layout=dict(xaxis=dict(title='Time'),
                       yaxis=dict(type='log',
                                  title='Fluxes (1/sec)'),
                       title='Latest predicted fluxes of reactions modeled by dFBA submodels'))
```
